```python
def check_input(sub_list):
    result = True

    if not sub_list:
        result = False
    elif len(sub_list) != 2:
        result = False
    else:
        time_str = sub_list[0]
        n = sub_list[1]

        if not isinstance(time_str, str):
            result = False
        elif time_str.count('.') != 1:
            result = False
        elif not isinstance(n, int):
            result = False
        elif n <= 0:
            result = False
        else:
            parts = time_str.split('.')

            if len(parts) != 2:
                result = False
            elif not (parts[0].isdigit() and parts[1].isdigit()):
                result = False
            else:
                hours = int(parts[0])
                minutes = int(parts[1])

                if not (0 <= hours <= 23 and 0 <= minutes <= 59):
                    result = False


    return result
```

```python
def calculate_time(valid_sublist):
    time_str = valid_sublist[0]
    n = valid_sublist[1]
    hours, minutes = time_str.split('.')
    hours = int(hours)
    minutes = int(minutes)

    new_hours = str(hours + n)
    new_minutes = str(minutes + n)

    if int(new_minutes) > 59:
        h = int(new_minutes) // 60
        m = int(new_minutes) % 60
        new_hours = str(int(new_hours) + h)
        new_minutes = str(m)
    if int(new_minutes) < 10:
        new_minutes = '0' + str(new_minutes)

    if int(new_hours) == 24:
        new_hours = '00'
    elif int(new_hours) > 23:
        new_hours = str(int(new_hours) - 24)
    if int(new_hours) < 10:
        new_hours = '0' + str(new_hours)

    new_time = '.'.join([new_hours, new_minutes])

    return new_time
```

```python
def main(user_input):
    final_time = ''

    for sub_list in user_input:

        if check_input(sub_list):
            valid_sublist = sub_list
            final_time = calculate_time(valid_sublist)

    print(final_time)
```

```python
main([['11', 8], ['14.02', -9], [], [5, '11.03'], ['11.03', 5]])
```
